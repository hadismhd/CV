```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import statsmodels.formula.api as smf
```

```python
df = pd.read_csv('animal_speeds.csv', sep=';')
pd.DataFrame(df)
```

```python
df = df[df['weight'] < 15000] 
pd.DataFrame(df)
```

```python
sns.lmplot(x='weight', y='highspeed', hue='movement_type', data=df, fit_reg=False, palette='viridis', size=7, aspect=2.5)
plt.ylabel("animal speed")
plt.xlabel("Animal Weight")
```

```python
model = smf.ols(formula='highspeed ~ weight + movement_type', data=df).fit()
summary = model.summary()
summary.tables[1]
```

```python
model_interaction = smf.ols(formula='highspeed ~ weight + movement_type + weight:movement_type', data=df).fit()
summary = model_interaction.summary()
summary.tables[1]
```

```python
x = df.drop(['highspeed','animal'], axis=1)
y = df['highspeed']
```

```python
# make sure it all numerical
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

categorical_features=['movement_type']
one_hot = OneHotEncoder()
transformer=ColumnTransformer([('one_hot',
                              one_hot,
                               categorical_features)],
                            remainder='passthrough' )
transformed_x = transformer.fit_transform(x)
pd.DataFrame(transformed_x)
```

```python
transformed_x = np.delete(transformed_x, 1, axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(transformed_x, y, test_size=0.3,random_state=0) 
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
linreg=LinearRegression()
model=linreg.fit(x_train,y_train)
```

```python
y_pred=linreg.predict(x_test)
y_pred
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
# plot the results
plt.figure(figsize=(15,10))
plt.scatter(y_test,y_pred)
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Actual vs Predicted')
```

```python
import seaborn as sns
sns.regplot(x=y_test,y=y_pred,ci=None,color ='red');
```

```python
fig, ax = plt.subplots(figsize=(10,4))

colors = {'climbing':'red', 'flying':'green', 'swimming':'blue', 'running':'purple'}


ax.scatter(df['weight'], df['highspeed'], c=df['movement_type'].map(colors))

plt.show()
```
